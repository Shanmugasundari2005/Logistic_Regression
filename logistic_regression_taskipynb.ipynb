## Logistic Regression

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/exam_scores (1).csv")
```

```python
print("Preview Dataset")
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
features = df.drop(columns='Pass',axis=1)
target = df['Pass']
```

```python
print(features)
```

```python
print(target)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(features)
```

```python
standardized_data=scaler.transform(features)
```

```python
features=standardized_data
target=df['Pass']
```

```python
print(features)
print(target)
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target,test_size = 0.2 ,random_state=42)
```

```python
print(features.shape,x_train.shape,x_test.shape)
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(x_train,y_train)
```

```python
y_prediction= classifier.predict(x_train)
training_data_accuracy = accuracy_score(y_train,y_prediction)
```

```python
print('Accuracy score of the training data:',training_data_accuracy)
```

```python
y_pre= classifier.predict(x_test)
testing_data_accuracy = accuracy_score(y_test,y_pre)
print(testing_data_accuracy)
```

```python
input_data = (52.59,52.69)
#changing into numpy array
input_data_as_numpy_array = np.asarray(input_data)
#reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
#standardize the input_data
std_data = scaler.transform (input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('Pass')
else:
 print('Fail')
```
